```python
import pandas as pd
import os
```

```python
os.listdir('../1_Data_scrapper/Laptopmedia_scrapper/Data')
```

```python
df = None

for dir in os.listdir('../1_Data_scrapper/Laptopmedia_scrapper/Data'):
    for file_name in os.listdir(f'../1_Data_scrapper/Laptopmedia_scrapper/Data/{dir}'):
        file_path = f'../1_Data_scrapper/Laptopmedia_scrapper/Data/{dir}/{file_name}'
        file = pd.read_csv(file_path)
        if df is None:
            df = file
        else:
            df = pd.concat([df, file], ignore_index=True)
```

```python
df.to_csv('data.csv', index_label=False)
```
